```python
import pandas as pd

# Read lenders data from CSV file into a DataFrame
lenders_df = pd.read_csv('/kaggle/input/mock-lenders1/mock_lenders.csv')

# Function to filter lenders based on client responses
def filter_lenders(client_responses, lenders_df):
    filtered_lenders_df = lenders_df.copy()
    for key, value in client_responses.items():
        # Remove non-numeric characters and convert to float
        filtered_lenders_df[key] = filtered_lenders_df[key].replace('[\$,]', '', regex=True).astype(float)
        filtered_lenders_df = filtered_lenders_df[filtered_lenders_df[key] <= value]
    return filtered_lenders_df

# Ask client questions and store responses
client_responses = {}
client_responses["minimum_credit_score"] = int(input("What is your minimum credit score? "))
client_responses["minimum_mrr"] = float(input("What is your minimum monthly recurring revenue? ").replace(',', ''))
client_responses["minimum_time_in_business"] = int(input("What is your minimum time in business (in years)? "))

# Filter lenders based on client responses
remaining_lenders_df = filter_lenders(client_responses, lenders_df)

# Display remaining lenders
if not remaining_lenders_df.empty:
    print("You are eligible for the following lenders:")
    print(remaining_lenders_df[['lender', 'interest_rate', 'loan_amount', 'logo']])
else:
    print("Sorry, no lenders match your criteria.")

```
